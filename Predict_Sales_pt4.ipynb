```python
# determine the average Item_Weight for each Item Type to replace nulls in column
avg_ItemWeight_filter = predict_sales_df.groupby('Item_Type').Item_Weight.mean()
avg_ItemWeight_filter
```

Interesting that the individual Item Weight is close when averaged.  I think its safe to replace the NaN Item Weight with Item Weight mean() value.

```python
# replace Height NaN values with mean() Weight

mean_Weight = predict_sales_df['Item_Weight'].mean()
predict_sales_df['Item_Weight'].fillna(value=mean_Weight, inplace=True)
predict_sales_df.head(15)
```

NaN Item Weight values replaced with mean()

```python
# replace Outlet_Size NaN with "Missing" string
predict_sales_df.fillna('Missing', inplace=True)
predict_sales_df.tail(15)
```

NaN Outlet Size values replaced with 'Missing' string

##Visualizations - Data Trends

```python
# created barplot to visually explain which items sold best in which stores/outlets.
# how OLD or NEW are the HIGH sales or LO sales stores?

# enlarged figure size for readability
plt.subplots(figsize=(10,10))

# plot multiple data variables on a single chart
sns.barplot(data=predict_sales_df, x='Outlet_Type', y='Item_Outlet_Sales', hue='Outlet_Establishment_Year')
plt.grid()
plt.title("Number of Sales per Store Type and Year Opened");
```


#Analysis

Supermarket Type 3 (1985 est) is the clear winner in sales($) for all the Outlet Types at 3500.

Grocery Stores (1985 and 1998 est) is the clear underperformer in sales for all the Outlet Types at about $300

Supermarket Types 2 & 1 (from 1987 to 2009 est) are closer with respect to Item Type sales ($2200 to 2500 range), but still underperform Supermarket Type 3 even though this store type has been around a long time.

#Recommendation
